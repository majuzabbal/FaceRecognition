```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Importar a imagem e transformar para a escala de cinza
imagem = cv2.imread("/home/maju/FaceRecognition/Todos/s01_13.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Visualizar imagem importada
plt.imshow(imagem, cmap='gray')
```

```python
# Classificador disponível no repositório oficial da OpenCV
classificador = cv2.CascadeClassifier("/home/maju/Downloads/haarcascade_frontalface_default.xml")
```

```python
# Utilizar o classifcador na imagem
faces = classificador.detectMultiScale(imagem, 1.3, 8)
```

```python
# Verificar a quantidade de faces encontradas na imagem
len(faces)
```

```python
# Posições dos rostos encontrados
faces[0]
```

```python
# Copiar a imagem para escrever na cópia
imagem_anotada = imagem.copy()
```

```python
# Escrever na cópia
for (x, y, w, h) in faces:
    cv2.rectangle(imagem_anotada, (x,y), (x+w, y+h), (255, 255, 0), 2)
```

```python
# Visualizar a cópia
plt.imshow(imagem_anotada, cmap='gray')
```

```python
# Cortar e salvar a imagem
for(x, y, w, h) in faces:
    imagem_roi = imagem[y:y+h, x:x+w]
    cv2.imwrite("s01_13.jpg", imagem_roi)
```
